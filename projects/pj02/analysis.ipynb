I Believe that there is a strong correlation between the percent of female primary education teachers and the percent of secondary education students that are female. Furthermore, I believe that a higher percentage of female primary education teachers will correspond to a higher percent of female secondary pupils.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
enrol_sec_female_per: str = "SE.SEC.ENRL.FE.ZS"
teach_prm_female: str = "SE.PRM.TCHR.FE.ZS"
```

The next cell will take a data csv file for World Bank data and read it into a DataFrame

```python
data_bank = pd.read_csv("../../data/databank_education_2018 (3).csv")
```

The next cell will print the shape of the data, as well as the first ten rows of it.

```python
print(data_bank.shape)
data_bank.head(10)
```

The next cell will give a specific set of columns corresponding to the three I want: country and data for my two factors, percent of female primary educators and percent of female secondary popuils enrolled.

```python
data_specified = data_bank[["Country", teach_prm_female, enrol_sec_female_per]]
data_specified.head(10)
```

The next cell will use row-based filtering to eliminate rows in which one of the two values has a "N/A" or "NaN". The function of the cell will be similar to the prior as it will print a certain number of rows for the two variables requested.

```python
not_na = data_bank[(data_bank[teach_prm_female].notna()) & (data_bank[enrol_sec_female_per].notna())]
print(not_na.head(5))
```

The next cell will sort the data in descending order and show the ten largest values of specified column in descending order.

```python
data_as_descend = data_bank.sort_values(by = [enrol_sec_female_per], ascending = False)
data_as_descend[["Country", teach_prm_female, enrol_sec_female_per]]
```

The next cell will create a scatter plot that shows the relationship for increase or decrease in percentage of females enrolled in secondary education as a function of percentage of primary educators that are female.

```python
var = data_as_descend.plot.scatter(teach_prm_female, enrol_sec_female_per, title = "Percentage of Female Primary Educators vs. Percentage of Female Students Enrolled in Secondary Education", xlabel = "Percentage of Female Primary Educators", ylabel = "Percentage of Female Students Enrolled in Secondary Education")
```

The last cell shows the correlation between the two variables analyzed: percentage of primary educators that are female and the effect (if any) on percentage of secondary education students that are female.

```python
data_bank[teach_prm_female].corr(data_bank[enrol_sec_female_per])
```

"Conclusion"
My initial hypothesis that a higher percentage of primary educators being female would lead to (or at least correlate) a higher percentage of females as enrolled secondary students. My basis for this hypothesis was that as a female, growing up with more females as your mentors and guides in education would lead to a higher chance of continuing education. However with a correlation of 0.33, this is not the case. I would attribute some of this to the fact that many developed countries have requirements for secondary education, so I would be interested to further this analysis in United Nation defined "developing" countries where this relationship may be stronger.
